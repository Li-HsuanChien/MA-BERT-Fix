```python
import os
import gdown
import zipfile
%cd MA-Bert
# Google Drive file ID for IMDB dataset
DATASET_URL = "https://drive.google.com/file/d/1Qlp3jBnonYJiu8Xqtp898gCbWopMC7nD/view?usp=sharing"
IMDB_DATASET_ZIPFILE_NAME = "corpus.zip"

# Directory where files will be saved
DIRECTORY = "."  # Directory where extracted files will go

# Path for downloaded .zip file
downloaded_file = os.path.join(DIRECTORY, IMDB_DATASET_ZIPFILE_NAME)

# Create the corpus directory if it doesn't exist
if not os.path.exists(DIRECTORY):
    os.makedirs(DIRECTORY)

# Download the .zip file if it doesn't exist
if not os.path.exists(downloaded_file):
    print(f"Downloading from: {DATASET_URL}")
    gdown.download(DATASET_URL, downloaded_file, quiet=False, fuzzy=True)
else:
    print(f"{downloaded_file} already exists, skipping download.")

# Check file size after download
if os.path.exists(downloaded_file):
    file_size = os.path.getsize(downloaded_file) / (1024 * 1024)  # Convert to MB
    print(f"Downloaded file size: {file_size:.2f} MB")

# Extract the .zip file directly into the corpus directory
print(f"Extracting: {downloaded_file}")
try:
    with zipfile.ZipFile(downloaded_file, 'r') as zip_ref:
        zip_ref.extractall(DIRECTORY)  # Extract all files into the corpus directory
    print(f"ZIP file extracted to {DIRECTORY}")
except zipfile.BadZipFile:
    print("Error: The downloaded file is not a valid zip file.")
```

```python
import os
import pandas as pd
pd.set_option('display.max_columns', None)
%cd /content/drive/MyDrive/MA-Bert
# Directory where the .tsv files are stored
DIRECTORY = "corpus"

# List of unzipped .tsv files
IMDB_DATASET_NAME = [
    "imdb/imdb.dev.txt.ss",
    "imdb/imdb.train.txt.ss",
    "imdb/imdb.train.txt.ss",
    "yelp_13/yelp-2013-seg-20-20.dev.ss",
    "yelp_13/yelp-2013-seg-20-20.test.ss",
    "yelp_13/yelp-2013-seg-20-20.train.ss",
    "yelp_14/yelp-2014-seg-20-20.dev.ss",
    "yelp_14/yelp-2014-seg-20-20.test.ss",
    "yelp_14/yelp-2014-seg-20-20.train.ss"
]

# Load each .tsv file into a pandas DataFrame and print a summary
for tsv_file in IMDB_DATASET_NAME:
    file_path = os.path.join(DIRECTORY, tsv_file)

    if os.path.exists(file_path):
        print(f"\nLoading {tsv_file} as pandas DataFrame")
        df = pd.read_csv(file_path, sep='\t', nrows=2)
        display(df)
    else:
        print(f"{file_path} does not exist.")
```
